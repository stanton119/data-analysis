## Get historic intensity for region

```python
# load from disk and from API if fails
# partition by month
import datetime

path_hist_int = data_path / "historic_intensity"
path_hist_int.mkdir(exist_ok=True)

postcode = "rg21"
start_dates = pd.date_range(
    start="2017-01-01", end=datetime.date.today() - datetime.timedelta(7), freq="1W-Mon"
)
# Needs 14 day range max...

for _idx in range(len(start_dates) - 1):
    date_from, date_to = start_dates[_idx].strftime("%Y-%m-%d"), start_dates[
        _idx + 1
    ].strftime("%Y-%m-%d")
    filepath = path_hist_int / date_from / date_to / f"{postcode}.json"
    print(f"Loading {filepath}")

    if filepath.exists():
        print("Already exists")
        continue

    filepath.parent.mkdir(parents=True, exist_ok=True)

    result_url = f"https://api.carbonintensity.org.uk/regional/intensity/{date_from}/{date_to}/postcode/{postcode}"
    result = requests.get(result_url)
    if result.status_code != 200:
        print(f"error for {result_url}")

    with open(filepath, "wb") as f:
        f.write(result.content)
```
